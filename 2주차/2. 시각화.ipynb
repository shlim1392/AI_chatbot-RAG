| Íµ¨Î∂Ñ | Î™©Ï†Å | ÎèÑÍµ¨ | ÌäπÏßï |

| **ÏõåÎìúÌÅ¥ÎùºÏö∞Îìú(Word Cloud)** | ÏûêÏ£º Îì±Ïû•Ìïú Îã®Ïñ¥Î•º ÌÅ¨Í∏∞Î°ú ÌëúÌòÑ | `wordcloud` | Îã®Ïñ¥ ÌÅ¨Í∏∞ = Îì±Ïû• ÌöüÏàò |
| **ÎßâÎåÄÍ∑∏ÎûòÌîÑ(Bar Plot)** | ÏÉÅÏúÑ NÍ∞ú Îã®Ïñ¥Î•º Ï†ïÎüâÏ†ÅÏúºÎ°ú ÎπÑÍµê | `matplotlib`, `pandas` | Ï†ïÌôïÌïú ÏàòÏπò ÎπÑÍµêÏóê Ïú†Ïö© |

```python
docs = [
    "ÎÇòÎäî Ïª§ÌîºÎ•º Ï¢ãÏïÑÌïúÎã§",
    "ÎÇòÎäî Ï∞®Î•º Ï¢ãÏïÑÌïúÎã§",
    "Ïª§ÌîºÎäî ÎßõÏûàÎã§",
    "Ï∞®Îäî Ìñ•Í∏∞Î°≠Îã§",
    "Ïª§ÌîºÎ•º Ï†ïÎßêÏ†ïÎßê Ï¢ãÏïÑÌïúÎã§"
]
```

```python
from collections import Counter


# Î™®Îì† Î¨∏Ïû•ÏùÑ Ìï©ÏπòÍ≥† Îã®Ïñ¥Î°ú ÎÇòÎàî
words = " ".join(docs).split() 


count = Counter(words)
count.most_common(10)
```

```python
import koreanize_matplotlib
from wordcloud import wordcloud
import matplotlib.pyplot as plt


font_path = "C:/Users//Downloads/NotoSansKR-Regular.ttf"
# WordCloud ÏÉùÏÑ±
wc = WordCloud(font_path=font_path,
               background_color="white",
               width=600, height=400,
               colormap="winter")

plt.figure(figsize=(8,6))
plt.imshow(wc.generate_from_frequencies(count)) # count_ ÎπàÎèÑÎ≥ÑÎ°ú Í∑∏Î¶º Í∑∏Î¶¨Í∏∞
plt.axis("off")
plt.show()

```

## **ÎßâÎåÄ Í∑∏ÎûòÌîÑ ÏãúÍ∞ÅÌôî Ïã§Ïäµ**

```python
import pandas as pd

df_top = pd.DataFrame(count.most_common(10), columns=['Îã®Ïñ¥', 'ÎπàÎèÑ'])
print(df_top)

plt.bar(df_top["Îã®Ïñ¥"], df_top["ÎπàÎèÑ"], color="gray")
plt.title("Top 10 Îã®Ïñ¥ ÎπÑÍµê")
plt.show()
```
