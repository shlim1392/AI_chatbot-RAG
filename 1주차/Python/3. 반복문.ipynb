
## breakÏôÄ continue
> Î∞òÎ≥µÎ¨∏(for, while)ÏóêÏÑú ÌùêÎ¶ÑÏùÑ Ï†úÏñ¥ÌïòÎäî Î™ÖÎ†πÏñ¥  




**break:**  
Î∞òÎ≥µÎ¨∏ÏùÑ Ï¶âÏãú Ï¢ÖÎ£å
(Ï°∞Í±¥ÏùÑ ÎßåÎÇòÎ©¥ Î∞òÎ≥µÏùÑ Î©àÏ∂§)  

**continue:**  
Ïù¥Î≤à Î∞òÎ≥µÎßå Í±¥ÎÑàÎõ∞Í≥†, Îã§Ïùå Î∞òÎ≥µÏúºÎ°ú ÎÑòÏñ¥Í∞ê  
(Ï°∞Í±¥ÏùÑ ÎßåÎÇòÎ©¥ ÏïÑÎûò ÏΩîÎìúÎ•º Ïã§ÌñâÌïòÏßÄ ÏïäÍ≥† Îã§Ïùå Î∞òÎ≥µ)

````python
for i in range(5):
    if i == 3:
        break  # iÍ∞Ä 3Ïù¥Î©¥ Î∞òÎ≥µÎ¨∏ Ï¢ÖÎ£å
    print(i)
````

````python
for i in range(5):
    if i == 3:
        continue  # iÍ∞Ä 3Ïù¥Î©¥ Í±¥ÎÑàÎõ∞Í≥† Îã§Ïùå Î∞òÎ≥µ
    print(i)
````
